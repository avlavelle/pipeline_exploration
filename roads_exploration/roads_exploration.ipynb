```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
interstates = gpd.read_file('Interstates/Interstates.shp')
```

```python
from shapely import wkt
```

```python
roads = pd.read_csv('roads.csv')
roads["geometry"] = roads['road_geometry'].apply(wkt.loads)
roads_gdf = gpd.GeoDataFrame(roads, geometry=roads['geometry'])
```

```python
roads_gdf.plot()
```

```python
roads_gdf.crs
```

```python
roads_gdf.crs
```

```python
interstates.plot()
```

```python
roads_gdf = roads_gdf.set_crs("EPSG:4326", inplace=True)
```

```python
roads_gdf = roads_gdf.to_crs(interstates.crs)
```

```python
ax = interstates.plot()
roads_gdf.plot(ax=ax, color='red', alpha=0.5)
```
